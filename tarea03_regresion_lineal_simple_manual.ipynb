## Parte 4: Método 3 - Descenso del Gradiente (Gradient Descent)

### Teoría:

El Descenso del Gradiente es un método de optimización **iterativo** utilizado para encontrar los parámetros que mejor ajustan un modelo de regresión. En lugar de calcular directamente los coeficientes mediante una fórmula cerrada (como en los mínimos cuadrados), este método aprende los valores óptimos de los parámetros moviéndose de manera gradual hacia el punto donde el error es mínimo.

El enfoque se basa en calcular el **gradiente**, es decir, la derivada de la función de costo. Este valor indica en qué dirección crece el error. Para minimizarlo, el algoritmo actualiza los parámetros moviéndose en la dirección contraria a ese gradiente.

La función de costo más común en regresión lineal es el **Error Cuadrático Medio (MSE)**:

$\text{MSE} = \frac{1}{n}\sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

### Proceso Iterativo Paso a Paso
El proceso sigue los siguientes pasos:

1. Comenzar con valores iniciales de $w_0$ y $w_1$
2. Calcular el error: $\text{MSE}$
3. Actualizar parámetros usando gradientes:

$$w_1 := w_1 - \alpha \frac{\partial MSE}{\partial w_1} = w_1 - \alpha \frac{-2}{n}\sum x_i(y_i - \hat{y}_i)$$

$$w_0 := w_0 - \alpha \frac{\partial MSE}{\partial w_0} = w_0 - \alpha \frac{-2}{n}\sum(y_i - \hat{y}_i)$$

4. Repetir hasta convergencia

Donde: 
- $\alpha$ es la **tasa de aprendizaje** (learning rate).
- Si $\alpha$ es muy grande, el algoritmo puede “saltarse” el mínimo.
- Si $\alpha$ es muy pequeño, convergerá correctamente, pero será más lento.

### Convergencia

La **convergencia** significa que el algoritmo ha llegado a un punto donde:
- los parámetros $w_0$ y $w_1$ ya no cambian de manera significativa,
- el error $\text{MSE}$ deja de disminuir entre iteraciones,
- y el modelo se encuentra cerca del mínimo de la función de costo.
